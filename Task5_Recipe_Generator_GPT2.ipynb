```python
# 🧠 Task 5: Recipe Generator using GPT-2

# Install dependencies
!pip install transformers torch --quiet

from transformers import pipeline, AutoTokenizer, AutoModelForSeq2SeqLM

# Load a text generation pipeline
generator = pipeline("text-generation", model="gpt2")

def generate_recipe(preferences, restrictions, max_length=300):
    """
    Generate a recipe based on user preferences and dietary restrictions.

    :param preferences: A list of user preferences (e.g., "spicy", "vegetarian").
    :param restrictions: A list of dietary restrictions (e.g., "gluten-free", "nut-free").
    :param max_length: The maximum length of the generated recipe.
    :return: A generated recipe.
    """
    # Create a prompt based on preferences and restrictions
    prompt = f"Generate a recipe that is "
    if preferences:
        prompt += " ".join([f"{pref}" for pref in preferences]) + ", "
    if restrictions:
        prompt += " ".join([f"{rest}" for rest in restrictions]) + ", "

    prompt = prompt.strip(', ')
    prompt += "."

    # Generate the recipe
    recipe = generator(prompt, max_length=max_length, num_return_sequences=1)[0]['generated_text']
    return recipe

# Example preferences and restrictions
preferences = ["spicy", "quick to prepare"]
restrictions = ["vegetarian", "gluten-free"]

# Generate the recipe
recipe = generate_recipe(preferences, restrictions)
print("Generated Recipe:", recipe)
```
