```python
# ðŸ§  Task 2: Machine Translation using MarianMT (English â†’ French)

# Install dependencies
!pip install transformers torch --quiet

from transformers import MarianMTModel, MarianTokenizer

# Function to load model and tokenizer
def load_model_and_tokenizer(model_name):
    # Load model and tokenizer
    model = MarianMTModel.from_pretrained(model_name)
    tokenizer = MarianTokenizer.from_pretrained(model_name)
    return model, tokenizer

# Example model for English to French translation
model_name = "Helsinki-NLP/opus-mt-en-fr"
model, tokenizer = load_model_and_tokenizer(model_name)

# Function to translate text
def translate_text(text, model, tokenizer):
    # Tokenize input text
    inputs = tokenizer(text, return_tensors="pt", padding=True)

    # Perform translation
    translated = model.generate(**inputs)

    # Decode the translated text
    translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)
    return translated_text

# Example usage
text_to_translate = "Hello, how are you?"
translated_text = translate_text(text_to_translate, model, tokenizer)
print(f"Translated text: {translated_text}")

# Interactive translation tool
def main():
    print("Welcome to the Translation Tool!")
    print("Available models:")
    print("1. English to French: Helsinki-NLP/opus-mt-en-fr")
    print()

    # Provide a list of available models and let the user choose
    available_models = {"1": "Helsinki-NLP/opus-mt-en-fr"}
    while True:
        model_choice = input("Enter the number of the model you want to use: ")
        if model_choice in available_models:
            model_name = available_models[model_choice]
            break
        else:
            print("Invalid choice. Please enter a valid number.")

    model, tokenizer = load_model_and_tokenizer(model_name)

    while True:
        text_to_translate = input("\nEnter text to translate (or type 'exit' to quit): ")
        if text_to_translate.lower() == 'exit':
            print("Goodbye! ðŸ‘‹")
            break

        translated_text = translate_text(text_to_translate, model, tokenizer)
        print(f"Translated text: {translated_text}")

# Run the translation tool
if __name__ == "__main__":
    main()
```
