```python
!pip install transformers diffusers torch --quiet
!pip install pillow --quiet
!pip install --upgrade --force-reinstall huggingface_hub
import torch
from PIL import Image
import matplotlib.pyplot as plt
from diffusers import StableDiffusionPipeline

model = StableDiffusionPipeline.from_pretrained("CompVis/stable-diffusion-v1-4", torch_dtype=torch.float16)
model = model.to("cuda")

def generate_oil_painting(prompt):
  output = model(prompt).images[0]
  return output

def display_image(output):
  plt.figure(figsize=(10, 7))
  plt.imshow(output)
  plt.axis("off")
  plt.show()

if __name__ == "__main__":
  prompt = "mountain landscape in oil painting style"
  output = generate_oil_painting(prompt)
  display_image(output)
```
